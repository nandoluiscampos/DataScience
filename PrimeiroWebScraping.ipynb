# Importando Blibliotecas

```python
# Importando o codigo
from bs4 import BeautifulSoup as bs
# Responsavel para fazer conexão http
import requests as rq
import re
```

# Pegando o html do site

```python
html_completo = rq.get("https://downloadcursos.top/")
html_formatado = bs(html_completo.content)
#print(html_formatado.prettify())
```

```python
# Pegar os h1 ou h2 em html
primeiro_caption = html_formatado.find("h2").text
print(primeiro_caption)
```

```python
# Pegar todos os h1 e h2 
todos_h2 = html_formatado.find_all(["h1" , "h2", "h3" ])
# colocar todos h1 em linhas separadas
for h2 in todos_h2:
    print(h2.text)
```

# Pegando Conteudos usando classes e IDs

```python
dados = html_formatado.find_all(class_="entry-title")
print (dados)
pre_requisito = html_formatado.find(class_='textwidget custom-html-widget')
print(pre_requisito)
```

# Pegando conteudo com filtro de texto

```python
todos_h2 = html_formatado.find_all(["h1" , "h2", "h3" ], text=re.compile("Abrindo"))
for h2 in todos_h2:
    print(h2.text)
todos_h2 = html_formatado.find_all(["div"])
for h2 in todos_h2:
    print(h2.text)    
```

# Pegar dados de um botão

```python
botao = html_formatado.find(class_="read-more-container").a
print(botao["href"])
```

```python
imagem = html_formatado.find("img")
print(imagem["src"])
```

# Pegar dados de uma lista

```python
elemento_lista = html_formatado.find("aside", id ="categories-2").li
print(elemento_lista)
```

```python
# . para classes e # para id
quarto_elemento_lista = html_formatado.select("#categories-2 ul li:nth-of-type(4)") 
print(quarto_elemento_lista )
```
